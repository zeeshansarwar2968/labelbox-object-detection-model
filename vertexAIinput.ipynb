### Dropping unnecessary columns

### Converting the nested values into a list for further transformations

```python
fd1_li = list(df1['labels'])
```

### Declaring lists to store the required values form the nested structure

```python
annotation_name = []
annotation_kind = []
bounding_box_top = []
bounding_box_left = []
bounding_box_height = []
bounding_box_width = []
classifications_name = []
classifications_type = []
```

### Looping through the list and appending the required values to the respective lists

```python
for i in range(len(fd1_li)):
    annotation_name.append(fd1_li[i][0]['annotations']['objects'][0]['name'])
    annotation_kind.append(fd1_li[i][0]['annotations']['objects'][0]['annotation_kind'])
    bounding_box_top.append(fd1_li[i][0]['annotations']['objects'][0]['bounding_box']['top'])
    bounding_box_left.append(fd1_li[i][0]['annotations']['objects'][0]['bounding_box']['left'])
    bounding_box_height.append(fd1_li[i][0]['annotations']['objects'][0]['bounding_box']['height'])
    bounding_box_width.append(fd1_li[i][0]['annotations']['objects'][0]['bounding_box']['width'])
    classifications_name.append(fd1_li[i][0]['annotations']['classifications'][0]['name'])
    classifications_type.append(fd1_li[i][0]['annotations']['classifications'][0]['radio_answer']['name'])
```

```python
df1.head()
```

### Adding the output to the dataframe

```python
df1['annotation_name'] = annotation_name
df1['annotation_kind'] = annotation_kind
df1['bounding_box_top'] = bounding_box_top
df1['bounding_box_left'] = bounding_box_left
df1['bounding_box_height'] = bounding_box_height
df1['bounding_box_width'] = bounding_box_width
df1['classifications_name'] = classifications_name
df1['classifications_type'] = classifications_type
```

```python
df1 = df1.drop(['labels'],axis=1)
```

Normalizing the bounding box vertices as required ny VertexAI import format`

```python
df3_columns = [ 'X_MIN', 'Y_MIN', 'X_MAX', 'Y_MAX']
```

```python
df3_min_max_scaled = df3_final.copy()
```

```python
# df3_min_max_scaled
```

```python
# apply normalization techniques
for column in df3_columns:
    df3_min_max_scaled[column] = (df3_min_max_scaled[column] - df3_min_max_scaled[column].min()) / (df3_min_max_scaled[column].max() - df3_min_max_scaled[column].min())    

```
